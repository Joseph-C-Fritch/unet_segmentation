```python
# Plotting Function
def image_show(image, nrows=1, ncols=1, cmap='gray'):
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(14, 14))
    ax.imshow(image, cmap='gray')
    ax.axis('off')
    return fig, ax
```

```python
# Function to read in Marks
def read_xml(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
    return lines
```

```python
# Get a list of files in Thyroid with Segmentatin Coordinates
filenames = glob.glob("thyroid/*")
```

```python
# Create List of Images Files
images = [x for x in filenames if x.endswith(".jpg")]

# Create List of Segments
segments = [x for x in filenames if x.endswith(".xml")]
```

```python
# Create List of Marks
marks = [read_xml(file) for file in segments]
```

```python
# Initialize a dictionary
temp_dict = {}

# Create a dictionary with img_id as the Key and the Mark as the Values
for i in marks:
    pattern = r"<case>(.*?)</case>"
    temp = re.findall(pattern, str(i))[0]
    
    id_1 = re.findall(r"<number>(.*?)</number>",temp)[0]
    id_2 = re.findall(r"<image>(.*?)</image>",temp)
    
    for t in id_2:
        pattern = r"<image>"+t+"</image><svg>(.*?)</svg>"
        temp_dict[id_1+'_'+t] = re.findall(pattern,temp)
```

```python
# Create DataFrame from Dictionary
segments = pd.DataFrame.from_dict(temp_dict, orient='index').reset_index()

# Rename columns
segments = segments.rename(columns = {'index':'img_id',0:"mark"})

# Look at df
segments.head()
```

```python
# Function to parse marks
def parse_mark(mark):
    i = 0
    t = 0
    test_1 = re.findall(r'\d+', mark)
    dims = len(test_1)//2
    temp = np.empty([dims, 2])
    
    while i < dims:
        temp[i] = (test_1[t], test_1[t+1])
        i = i+1
        t = t+2
        
    return temp
```

```python
# Parse Marks
df_new['mark'] = df_new['mark'].apply(lambda x: parse_mark(x))
```

```python
#Augment Images

import random
from scipy import ndarray
import skimage as sk
from skimage import transform
from skimage import util
import os

# our folder path containing some images
folder_path_i = "data/images/"
folder_path_m = "data/masks_inverted/"
# the number of file to generate
num_files_desired = 1500
#Random transformation choices
choice_list = [1,2,3]

# loop on all files of the folder and build a list of files paths
images = glob.glob("data/images/*")
images = [x.replace('data/images/','') for x in images]

num_generated_files = 0
while num_generated_files <= num_files_desired:
    # random image from the folder
    image_name = random.choice(images)
    # read image as an two dimensional array of pixels
    image_to_transform = sk.io.imread(folder_path_i+image_name)
    mask_to_transform = sk.io.imread(folder_path_m+image_name)
    num_generated_files = num_generated_files + 1
    
    # random num of transformations to apply
    num_of_trans = random
    num_transformations = 0
    transformed_image = None
    transformed_mask = None
    while num_transformations <= 1:
        # choose a random transformation to apply for a single image and mask
        trans_type = random.choice(choice_list)
        if trans_type > 2:     #Rotate Only 
            random_degree = random.uniform(-20, 20)
            transformed_image = sk.transform.rotate(image_to_transform, random_degree)
            transformed_mask = sk.transform.rotate(mask_to_transform, random_degree)
        elif trans_type > 1:  #Horizonal Flip Only 
            transformed_image = image_to_transform[:, ::-1]
            transformed_mask = mask_to_transform[:, ::-1]
        else:                  #Rotate and Horizonal Flip
            random_degree = random.uniform(-20, 20)
            transformed_image = sk.transform.rotate(image_to_transform, random_degree)
            transformed_mask = sk.transform.rotate(mask_to_transform, random_degree)
            transformed_image = transformed_image[:, ::-1]
            transformed_mask = transformed_mask[:, ::-1]
                       
        num_transformations += 1
    
    # define a name for our new file
    new_file_path_i = '%s/augmented_image_%s.jpg' % (folder_path_i, num_generated_files)
    new_file_path_m = '%s/augmented_image_%s.jpg' % (folder_path_m, num_generated_files)
    # write image to the disk
    sk.io.imsave(new_file_path_i, transformed_image)
    sk.io.imsave(new_file_path_m, transformed_mask)

```
